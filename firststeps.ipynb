```python
img_1 = imread(r"data\Otsu_data\N2DH-GOWT1\gt\man_seg01.tif") 
figure()  
imshow(img_1, 'gray')
```

```python
copy = img_1.copy()
import scipy.ndimage
median = scipy.ndimage.median_filter(copy, 5)
figure()
imshow(median, 'gray')
colorbar()

```

```python

def histogramm (img):
    result = list()

    for p in ndindex(img.shape):
    
        result.append(img[p])
       
    hist(result, bins = 50, density=True, 
                            edgecolor="#6A9662",
                            color="#DDFFDD")


histogramm(img_1)
plt.xlim([0,50])
```
