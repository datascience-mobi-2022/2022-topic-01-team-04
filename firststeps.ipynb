```python
%pylab
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img_1 = imread(r"data\Otsu_data\NIH3T3\gt\0.png") 
figure()  
imshow(img_1, 'gray')


img_2 = imread(r"data\Otsu_data\NIH3T3\img\dna-0.png") 
figure()    
imshow(img_2, 'gray')

img_3 = imread(r"data\Otsu_data\N2DH-GOWT1\img\t01.tif") 
figure()  
imshow(img_3, 'gray')


img_4 = imread(r"data\Otsu_data\N2DL-HeLa\img\t13.tif") 
figure()    
imshow(img_4, 'gray')

```

```python
copy = img_2.copy()
import scipy.ndimage
median = scipy.ndimage.median_filter(copy, 5)
figure()
imshow(median, 'gray')
colorbar()

gauss = scipy.ndimage.gaussian_filter(copy, 5)
figure()
imshow(gauss, 'gray')
colorbar()

gauss2 = scipy.ndimage.gaussian_filter(img_3, 5)
figure()
imshow(gauss2, 'gray')
colorbar()

gauss3 = scipy.ndimage.gaussian_filter(img_4, 5)
figure()
imshow(gauss3, 'gray')
colorbar()

median2 = scipy.ndimage.median_filter(img_4, 5)
figure()
imshow(median2, 'gray')
colorbar()


```

```python
plt.hist(median2.ravel(), density=True, bins=10000, log = 'True', fc='b', ec='b')

```
