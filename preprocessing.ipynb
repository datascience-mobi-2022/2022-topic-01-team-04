## Preprocessing algorithms
### Histogram stretching
**current runtime ~ 6.6 s/picture**

- **Trial** cell 5
- _Saving the processed images automatically is not yet implemented, the algorithm creates a local copy of the original image, therefore does not influence the original picture_

```python
# histogram stretching algorithm

def stretch(x):
    intensities = []
    img = copy(images[x])
    h = img.shape[0]
    w = img.shape[1]
    for x, y in np.ndindex((h,w)):
        intensities.append(img[x,y])
    img_max = max(intensities)
    img_min = min(intensities)
    for a, b in np.ndindex((h,w)):
        img[a,b] = (img[a,b]-img_min)*256/(img_max-img_min)
    return img 
```
