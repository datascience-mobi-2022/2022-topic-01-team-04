```python
#histogramm der ausgewählten Bilder der Präsentation zur Beschreibung von Challenge und Lösungsansatz
#für NIH3T3
imgdna44 = imread (r"data\Otsu_data\NIH3T3\img\dna-44.png")
figure()
imshow(imgdna44, 'gray')
colorbar()

```
