```python
%pylab
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img_1 = imread(r"data\Otsu_data\N2DH-GOWT1\img\t39.tif") 
figure()  
imshow(img_1, 'gray')

img_1_gt = imread(r"data\Otsu_data\N2DH-GOWT1\gt\man_seg39.tif") 
figure()  
imshow(img_1_gt, 'gray')
```

```python
plt.hist(img_1.flatten(), bins=10000, log = 'True', fc='b', ec='b')
```

```python
import scipy.ndimage
gauss1 = scipy.ndimage.gaussian_filter(img_1,13)
figure()
imshow(gauss1, 'gray')
colorbar()
```

```python
plt.hist(gauss1.flatten(), bins = 1000, log = 'True', ec = 'b', fc = 'b')
```

```python
plt.hist(img_1_gt.flatten(), bins=10000, log = 'True', fc='b', ec='b')
```

```python
median1 = scipy.ndimage.median_filter(gauss1, 11)
figure()
imshow(median1, 'gray')
colorbar()
```
